```python
data = pd.read_csv('/drive/MyDrive/TriceMBA.csv')
```

```python
data.head()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
basket_hot = (data
          .groupby(['UID', 'Service'])['Count']
          .sum().unstack().reset_index().fillna(0)
          .set_index('UID'))
```

```python
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
      
```

```python
basket_encoded = basket_hot.applymap(hot_encode)
basket_hot = basket_encoded
```

```python
basket_hot
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frq_items = apriori(basket_hot, min_support = 0.15, use_colnames = True)
frq_items
```

```python
pd.set_option('display.max_columns', 10)
```

```python
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head())
```

```python
rules.to_excel('MBA_output.xlsx')
```
